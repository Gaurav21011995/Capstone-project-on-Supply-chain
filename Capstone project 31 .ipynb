Running condition and assigned as x variale for India to List all the cities and the respective number of pending orders in 'India' with crosstab function

```python
x = df[df['Order Country']=='India']
```

```python
x
```

```python
pd.crosstab(x['Order City'],x['Order Status'])[['PENDING']]
```

Running crosstab function again to find Which country has the most suspected fraud

```python
pd.crosstab(df['Order Country'],df['Order Status'])[['SUSPECTED_FRAUD']].sort_values('SUSPECTED_FRAUD')
```
